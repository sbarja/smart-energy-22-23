# Estadística 

```python
import pandas as pd #importamos Pandas!
import matplotlib.pyplot as plt #importamos matplotlib para hacer gráficos
import seaborn as sns #importamos seaborn para hacer gráficos
```

```python
# Importamos Excel
consumos = pd.read_excel('Data/data-consumo-usuarios-oversampling.xlsx')
consumos.head()
```

```python
# Buscamos valores en cluster
consumos['cluster'].unique()
```

```python
# Buscamos número de CUPs/clientes
consumos['CUPs'].nunique()
```

Forma

```python
# Forma del dataset (444 filas, 26 columnas)
# Las columnas son (1) CUPs, (1) cluster y (24) consumo horario (de h-0 a h-23)
consumos.shape
```

```python
consumos.describe()
```

```python
# Miramos si hay algun valor NaN
consumos.isna().sum()
```

```python
# Vamos a rellenar los NaN
consumos_transposed = consumos.T.interpolate(method='linear')
```

```python
consumos = consumos_transposed.T
consumos.isna().sum()
```

```python
# Vamos a ver cuantos casos tenemos en cada uno de los clusters. Tenemos un dataset balanceado?
# Casos cluster==0
print("Casos cluster 0:", consumos[consumos['cluster'] == 0]['cluster'].count())
# Casos cluster=1
print("Casos cluster 1:", consumos[consumos['cluster'] == 1]['cluster'].count())
```

```python
# Creamos dos dataframes (uno por cluster) para analizarlos por separado
clientes_0 = consumos.loc[consumos['cluster'] == 0]
clientes_1 = consumos.loc[consumos['cluster'] == 1]
```

```python
# Comparacion consumo horario promedio
print("Potencia horaria media cluster 0: ", clientes_0.drop(['CUPs','cluster'], axis=1).mean(axis=1).mean(), 'kW')
print("Potencia horaria media cluster 1: ", clientes_1.drop(['CUPs','cluster'], axis=1).mean(axis=1).mean(), 'kW')
```

```python
# Eliminamos la columna 'cluster' para poder hacer un plot de las distintas curvas de carga
df_0 = clientes_0.drop(['cluster'], axis=1)
# Hacemos que la columna CUPs sea el indice (tiene sentido, ya que cada fila tiene un valor distinto e identifica el SM)
df_0.set_index(['CUPs'], inplace=True)
# Transponemos la matriz, por facilidad de hacer el plot
df_0 = df_0.T
# Cambiamos el nombre del índice a "hour"
df_0.index.name = 'hour'
df_0.head()
```

```python
df_1 = clientes_1.drop(['cluster'], axis=1)
df_1.set_index(['CUPs'], inplace=True)
df_1 = df_1.T
df_1.index.name = 'hour'
```

```python
# Obtenemos una lista con las columnas de los dos dfs para tener los CUPs del cluster 0 y del cluster 1
cups_0 = df_0.columns
cups_1 = df_1.columns

# Creamos figura (figsize indica el tamaño)
plt.figure(figsize=(20,8))
# Creamos un bucle donde cups vaya cogiendo cada uno de los strings dentro de la lista cups_0
for cups in cups_0:
    # 'lightcoral' indica el color (aquí hay más colores: https://matplotlib.org/2.1.1/gallery/color/named_colors.html)
    # linewidth fija el grosor de la línia y alpha la transparencia (más cercano a 0, más transparente)
    plt.plot(df_0[cups], 'lightcoral', linewidth=1, alpha=0.4)
for cups in cups_1:
    plt.plot(df_1[cups], 'green', linewidth=1, alpha=0.4)

# Queremos que el eje x tenga marcadas las horas
plt.xticks(df_0.index)
# Fíjamos los márgenes a x=0 e y=0
plt.margins(x=0, y=0)
plt.show()  
```

```python
# Añadimos consumo medio para distinguir con más claridad las diferencias entre los clusters
df_0['mean'] = df_0.mean(axis=1)
df_1['mean'] = df_1.mean(axis=1)
df_0.head()
```

```python
# Creamos los mismos gráficos que antes, añadiendo las curvas promedio de los dos clusters con más opacidad (alpha)
plt.figure(figsize=(20,8))
for cups in cups_0:
    plt.plot(df_0[cups], 'lightcoral', linewidth=1, alpha=0.2)

for cups in cups_1:
    plt.plot(df_1[cups], 'green', linewidth=1, alpha=0.2)

plt.plot(df_0['mean'], 'lightcoral', linewidth=4, alpha=1)    
plt.plot(df_1['mean'], 'green', linewidth=4, alpha=1)

plt.xticks(df_0.index)
plt.margins(x=0, y=0)
plt.show()  
```

```python
# Vamos a ver la correlación entre features (o columnas)
# Creamos figura
plt.figure(figsize=(10, 8))
# Creamos matriz de correlaciones después de eliminar columna CUPs ya que no nos ofrece información en este caso
corr = consumos.drop(['CUPs'],axis=1).corr()
# Creamos un mapa de calor para detectar visualmente la correlacion entre las columnas
sns.heatmap(corr, cmap="coolwarm")
```

```python
# Se puede ver que el consumo entre las 13h y las 21h tiene una correlación mayor con el cluster que el consumo entre 6h y 12h
# Mirando las curvas de carga, vemos que por la mañana se parecen bastante, mientras que durante la noche la diferencia aumenta
```

```python
# Vamos a crear ahora algunos Boxplots para detectar la variabilidad dentro de cada cluster
plt.subplots(figsize=(15, 8))
# Creating plot
bp = clientes_0.drop(['CUPs'],axis=1).boxplot(column=list(clientes_0.drop(['CUPs'],axis=1).columns))
# show plot
plt.show()
```

```python
plt.subplots(figsize=(15, 8))
# Creating plot
bp = clientes_1.drop(['CUPs'],axis=1).boxplot(column=list(clientes_1.drop(['CUPs'],axis=1).columns))
# show plot
plt.show()
```

```python
# Vamos a crear algunos nuevos features que puedan ser interesantes para reducir la dimensionalidad del problema
# y mejorar el rendimiento del algoritmo

# Crearemos nuevos features partiendo del consumo horario (mean, max, std, mean(13h-21h)), así que creamos una lista de las 
# horas para poder hacer un slicing del dataframe
horas = list(consumos.drop(['CUPs', 'cluster'], axis=1))

# Ejemplos básicos (ojo, puede que algunas de estas características tengan una alta correlación entre ellas)
consumos['promedio'] = consumos[horas].mean(axis=1)
consumos['maximo'] = consumos[horas].max(axis=1)
consumos['minimo'] = consumos[horas].min(axis=1)
consumos['desvio'] = consumos[horas].std(axis=1)

# Ejemplo minmax
minmax = []
# iteramos fila a fila en nuestro df
for index, row in consumos.iterrows():
    # si el mínimo es 0, fijaremos minmax a 0, para evitar una indeterminación 0/0
    if row['minimo'] == 0:
        minmax.append(0)
    else:
        minmax.append(row['minimo']/row['maximo'])
consumos['minmax'] = minmax

# Ejemplo promedio durante un periodo de tiempo. Hemos visto que entre 13h y 21h hay una mayor diferencia entre clusters. 
horas_puntas = ['h-' + str(x) for x in range(13,21)]
consumos['promedio_punta'] = consumos[horas_puntas].mean(axis=1)

consumos.head()
```

```python
# Volvamos a ver la correlación
f, ax = plt.subplots(figsize=(10, 8))
corr = consumos.drop(['CUPs'],axis=1).corr()
sns.heatmap(corr, cmap="coolwarm")
# Correlación negativa (cercana a -1) también es interesante, como puede ser el caso de minmax y cluster
```
